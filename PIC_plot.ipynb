```python
##45 degree middle
import matplotlib.pyplot as plt
#file import
file = 'middle45.txt'
time,Y = [],[]
with open(file, 'r',encoding="utf-8") as data:
    lines = data.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        time.append(value[0])  #time(x)
        Y.append(value[1])  #flight path(y)

plt.plot(time, Y)
plt.title("Launch at a 45 degree angle", fontsize=16) #chart_topic
plt.xlabel("time", fontsize=12) #x_label
plt.ylabel("height (flight path) ", fontsize=12) #y_label

#plt.savefig('45度中杯子發射.png', bbox_inches='tight')
#plt.show()

##90 degree middle
import matplotlib.pyplot as plt

#file import
file_90 = 'middle90.txt'
X_90,Y_90 = [],[]
with open(file_90, 'r',encoding="utf-8") as data_90:
    lines_90 = data_90.readlines()
    for line in lines_90:
        value = [float(s) for s in line.split()]
        X_90.append(value[2])  #time(x)
        Y_90.append(value[1])  #flight path(y)

plt.plot(X_90, Y_90)
plt.title("Launch at a 90 degree angle", fontsize=16) #chart_topic
plt.xlabel("x", fontsize=12) #x_label
plt.ylabel("y", fontsize=12) #y_label

plt.savefig('90度中杯子發射.png', bbox_inches='tight')
plt.show()
```
