```python
##shape compare(height)
import matplotlib.pyplot as plt

## 45 middle file import
file_45 = 'middle45.txt'
time_45,y_45,X_45 = [],[],[]
with open(file_45, 'r',encoding="utf-8") as data_45:
    lines_45 = data_45.readlines()
    for line in lines_45:
        value = [float(s) for s in line.split()]
        time_45.append(value[0])  #time
        y_45.append(value[1])  #flight path(y)
        X_45.append(value[2])  #flight path(x)

## large file import  
file_l = 'large.txt'
time_l,y_l,X_l = [],[],[]
with open(file_l, 'r',encoding="utf-8") as data_l:
    lines_l = data_l.readlines()
    for line in lines_l:
        value = [float(s) for s in line.split()]
        time_l.append(value[0])  #time
        y_l.append(value[1])  #flight path(y)
        X_l.append(value[2])  #flight path(x)
        
## small file import  
file_s = 'small.txt'
time_s,y_s,X_s = [],[],[]
with open(file_s, 'r',encoding="utf-8") as data_s:
    lines_s = data_s.readlines()
    for line in lines_s:
        value = [float(s) for s in line.split()]
        time_s.append(value[0])  #time
        y_s.append(value[1])  #flight path(y)
        X_s.append(value[2])  #flight path(x)
        
##data plot(xt)
line_45, = plt.plot(time_45,y_45,label='middle cup')
line_s, = plt.plot(time_s,y_s,label='small cup')
line_l, = plt.plot(time_l,y_l,label='large cup')

##chart set
plt.title("Comparison of different shape cup", fontsize=16) #chart_topic
plt.xlabel("time", fontsize=12) #x_label
plt.ylabel("y (m)", fontsize=12) #y_label

##ledend and
plt.legend(handles=[line_45, line_s, line_l])

##show
#plt.savefig('不同杯子大小發射距離yt圖.png', bbox_inches='tight')
plt.show()
```
