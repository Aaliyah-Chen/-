```python
##45 degree middle
import matplotlib.pyplot as plt
import numpy as np
#file import
file = 'middle45.txt'
x,Y = [],[]
with open(file, 'r',encoding="utf-8") as data:
    lines = data.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        x.append(-value[2])  #time(x)
        Y.append(value[1])  #flight path(y)
        

plt.plot(x, Y,'.',markersize = '5')
plt.axvline(-0.7, linestyle= '--',color= 'c')
plt.axvline(0, linestyle= '--',color= 'c')
plt.annotate('Ⅰ', xy=(-1.5, -1),fontsize=24)
plt.annotate('ⅠⅠ', xy=(-0.5, -1),fontsize=24)
plt.annotate('ⅠⅠⅠ',xy=(1, -1),fontsize=24)



plt.title("Launch at a 45 degree angle", fontsize=16) #chart_topic
plt.xlabel("X (m)", fontsize=12) #x_label
plt.ylabel("Y (m)", fontsize=12) #y_label

plt.savefig('45度中杯子發射.png', bbox_inches='tight')

plt.show()
```

```python
##45 degree middle
import matplotlib.pyplot as plt
#file import
file = '45.txt'
time,v = [],[]
with open(file, 'r',encoding="utf-8") as data:
    lines = data.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        time.append(value[0])  #time(x)
        v.append(value[1])  #flight path(y)

plt.plot(time, v)
plt.title("Launch at a 45 degree angle", fontsize=16) #chart_topic
plt.xlabel("time", fontsize=12) #x_label
plt.ylabel("Velocity (m/s)", fontsize=12) #y_label

plt.savefig('vt.png', bbox_inches='tight')
plt.show()
```
